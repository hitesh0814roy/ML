```python
class OpenSkyTracker:
    def __init__(self):
        self.base_url = "https://opensky-network.org/api"

    def get_flights_by_airport(self, airport: str, begin: int, end: int) -> Optional[Dict]:
        """
        Get flights by airport within time range
        begin and end are Unix timestamps
        """
        endpoint = f"{self.base_url}/flights/airport"
        params = {
            'icao': airport,
            'begin': begin,
            'end': end
        }

        try:
            response = requests.get(endpoint, params=params)
            response.raise_for_status()
            return response.json()
        except requests.exceptions.RequestException as e:
            print(f"Error fetching data: {e}")
            return None

    def get_all_flights(self) -> Optional[Dict]:
        """
        Get all current flights (limited data)
        """
        endpoint = f"{self.base_url}/states/all"

        try:
            response = requests.get(endpoint)
            response.raise_for_status()
            return response.json()
        except requests.exceptions.RequestException as e:
            print(f"Error fetching data: {e}")
            return None
```

```python
# Install required packages


print(" Flight Tracker with Live Maps - Ready to use!")

class FlightTracker:
    def __init__(self, api_key: str):
        self.api_key = api_key
        self.base_url = "http://api.aviationstack.com/v1"

    def get_flights_by_airline(self, airline_iata: str, limit: int = 10) -> Optional[Dict]:
        """Get flights by airline IATA code"""
        endpoint = f"{self.base_url}/flights"
        params = {
            'access_key': self.api_key,
            'airline_iata': airline_iata,
            'limit': limit
        }

        try:
            response = requests.get(endpoint, params=params)
            response.raise_for_status()
            return response.json()
        except requests.exceptions.RequestException as e:
            print(f" Error fetching data: {e}")
            return None

    def get_flights_by_flight_number(self, flight_number: str) -> Optional[Dict]:
        """Get flight by flight number"""
        endpoint = f"{self.base_url}/flights"
        params = {
            'access_key': self.api_key,
            'flight_iata': flight_number
        }

        try:
            response = requests.get(endpoint, params=params)
            response.raise_for_status()
            return response.json()
        except requests.exceptions.RequestException as e:
            print(f" Error fetching data: {e}")
            return None

    def get_flights_by_airport(self, airport_iata: str, flight_type: str = "departure") -> Optional[Dict]:
        """Get flights by airport (departure or arrival)"""
        endpoint = f"{self.base_url}/flights"
        params = {
            'access_key': self.api_key,
            f'{flight_type}': airport_iata,
            'limit': 20
        }

        try:
            response = requests.get(endpoint, params=params)
            response.raise_for_status()
            return response.json()
        except requests.exceptions.RequestException as e:
            print(f" Error fetching data: {e}")
            return None

    def display_flight_info(self, flight_data: Dict):
        """Display formatted flight information"""
        if not flight_data or 'data' not in flight_data or not flight_data['data']:
            print(" No flight data available")
            return None

        flights_info = []
        for flight in flight_data['data']:
            flight_info = {
                'flight_number': flight.get('flight', {}).get('iata', 'N/A'),
                'airline': flight.get('airline', {}).get('name', 'N/A'),
                'departure_airport': flight.get('departure', {}).get('airport', 'N/A'),
                'departure_iata': flight.get('departure', {}).get('iata', 'N/A'),
                'arrival_airport': flight.get('arrival', {}).get('airport', 'N/A'),
                'arrival_iata': flight.get('arrival', {}).get('iata', 'N/A'),
                'status': flight.get('flight_status', 'N/A'),
                'scheduled_departure': flight.get('departure', {}).get('scheduled'),
                'estimated_departure': flight.get('departure', {}).get('estimated'),
                'scheduled_arrival': flight.get('arrival', {}).get('scheduled'),
                'estimated_arrival': flight.get('arrival', {}).get('estimated')
            }
            flights_info.append(flight_info)

            print("\n" + "="*60)
            print(f"截  Flight: {flight_info['flight_number']}")
            print(f" Airline: {flight_info['airline']}")
            print(f" From: {flight_info['departure_airport']} ({flight_info['departure_iata']})")
            print(f" To: {flight_info['arrival_airport']} ({flight_info['arrival_iata']})")

            status = flight_info['status']
            status_emoji = "" if status == 'active' else "" if status == 'scheduled' else ""
            print(f" Status: {status_emoji} {status}")

            print(f" Scheduled Departure: {self.format_time(flight_info['scheduled_departure'])}")
            print(f" Estimated Departure: {self.format_time(flight_info['estimated_departure'])}")
            print(f" Scheduled Arrival: {self.format_time(flight_info['scheduled_arrival'])}")
            print(f" Estimated Arrival: {self.format_time(flight_info['estimated_arrival'])}")

        return flights_info

    def format_time(self, time_str: str) -> str:
        """Format time string for better readability"""
        if not time_str:
            return "N/A"
        try:
            return time_str.replace('T', ' ').split('+')[0]
        except:
            return time_str

class OpenSkyTracker:
    def __init__(self):
        self.base_url = "https://opensky-network.org/api"

    def get_all_flights(self) -> Optional[Dict]:
        """Get all current flights (limited data)"""
        endpoint = f"{self.base_url}/states/all"

        try:
            response = requests.get(endpoint)
            response.raise_for_status()
            return response.json()
        except requests.exceptions.RequestException as e:
            print(f" Error fetching data: {e}")
            return None

    def get_flight_track(self, icao24: str) -> Optional[Dict]:
        """Get track for a specific flight"""
        endpoint = f"{self.base_url}/tracks"
        params = {'icao24': icao24}

        try:
            response = requests.get(endpoint, params=params)
            response.raise_for_status()
            return response.json()
        except requests.exceptions.RequestException as e:
            print(f" Error fetching track: {e}")
            return None

class FlightMapVisualizer:
    def __init__(self):
        self.airport_coordinates = self.load_airport_coordinates()

    def load_airport_coordinates(self) -> Dict:
        """Load common airport coordinates"""
        return {
            'JFK': (40.6413, -73.7781),
            'LAX': (33.9416, -118.4085),
            'SFO': (37.6213, -122.3790),
            'ORD': (41.9742, -87.9073),
            'DFW': (32.8998, -97.0403),
            'ATL': (33.6407, -84.4277),
            'LHR': (51.4700, -0.4543),
            'CDG': (49.0097, 2.5479),
            'DXB': (25.2532, 55.3657),
            'HND': (35.5494, 139.7798),
            'SYD': (-33.9399, 151.1753),
            'DEL': (28.5562, 77.1000)
        }

    def get_airport_coords(self, airport_code: str) -> Tuple[float, float]:
        """Get coordinates for airport, with fallback to random nearby location"""
        if airport_code in self.airport_coordinates:
            return self.airport_coordinates[airport_code]
        else:
            # Generate random coordinates near a major airport for demo
            base_lat, base_lon = random.choice(list(self.airport_coordinates.values()))
            return (base_lat + random.uniform(-2, 2), base_lon + random.uniform(-2, 2))

    def create_flight_path_map(self, flights_info: List[Dict]) -> folium.Map:
        """Create a map with flight paths"""
        if not flights_info:
            print(" No flight data to display on map")
            return None

        # Calculate center of map
        lats = []
        lons = []
        for flight in flights_info:
            dep_coords = self.get_airport_coords(flight['departure_iata'])
            arr_coords = self.get_airport_coords(flight['arrival_iata'])
            lats.extend([dep_coords[0], arr_coords[0]])
            lons.extend([dep_coords[1], arr_coords[1]])

        center_lat = sum(lats) / len(lats)
        center_lon = sum(lons) / len(lons)

        # Create map
        m = folium.Map(
            location=[center_lat, center_lon],
            zoom_start=3,
            tiles='OpenStreetMap'
        )

        # Add flight paths
        for i, flight in enumerate(flights_info):
            dep_coords = self.get_airport_coords(flight['departure_iata'])
            arr_coords = self.get_airport_coords(flight['arrival_iata'])

            # Create great circle path (simplified)
            path = self.create_great_circle_path(dep_coords, arr_coords)

            # Choose color based on status
            color = 'green' if flight['status'] == 'active' else 'blue' if flight['status'] == 'scheduled' else 'red'

            # Add flight path
            folium.PolyLine(
                path,
                color=color,
                weight=3,
                opacity=0.7,
                popup=f"Flight {flight['flight_number']}: {flight['departure_iata']}  {flight['arrival_iata']}"
            ).add_to(m)

            # Add departure marker
            folium.Marker(
                dep_coords,
                popup=f" Departure: {flight['departure_airport']} ({flight['departure_iata']})",
                tooltip=f"Departure: {flight['departure_iata']}",
                icon=folium.Icon(color='green', icon='plane', prefix='fa')
            ).add_to(m)

            # Add arrival marker
            folium.Marker(
                arr_coords,
                popup=f" Arrival: {flight['arrival_airport']} ({flight['arrival_iata']})",
                tooltip=f"Arrival: {flight['arrival_iata']}",
                icon=folium.Icon(color='red', icon='plane', prefix='fa')
            ).add_to(m)

            # Add moving plane icon (simulated)
            if len(path) > 1:
                mid_point = path[len(path)//2]
                folium.Marker(
                    mid_point,
                    popup=f"截 Flight {flight['flight_number']} - {flight['airline']}",
                    tooltip=f"Flight {flight['flight_number']}",
                    icon=folium.Icon(color='blue', icon='plane', prefix='fa')
                ).add_to(m)

        # Add layer control
        folium.LayerControl().add_to(m)

        return m

    def create_great_circle_path(self, point1: Tuple[float, float], point2: Tuple[float, float], num_points: int = 50) -> List[Tuple[float, float]]:
        """Create a great circle path between two points"""
        lat1, lon1 = math.radians(point1[0]), math.radians(point1[1])
        lat2, lon2 = math.radians(point2[0]), math.radians(point2[1])

        path = []
        for i in range(num_points + 1):
            fraction = i / num_points

            # Spherical interpolation
            d = math.acos(math.sin(lat1) * math.sin(lat2) + math.cos(lat1) * math.cos(lat2) * math.cos(lon2 - lon1))

            if d == 0:
                # Points are the same
                path.append(point1)
                continue

            a = math.sin((1 - fraction) * d) / math.sin(d)
            b = math.sin(fraction * d) / math.sin(d)

            x = a * math.cos(lat1) * math.cos(lon1) + b * math.cos(lat2) * math.cos(lon2)
            y = a * math.cos(lat1) * math.sin(lon1) + b * math.cos(lat2) * math.sin(lon2)
            z = a * math.sin(lat1) + b * math.sin(lat2)

            lat = math.atan2(z, math.sqrt(x**2 + y**2))
            lon = math.atan2(y, x)

            path.append((math.degrees(lat), math.degrees(lon)))

        return path

    def create_realtime_flight_map(self, flight_data: Dict) -> folium.Map:
        """Create real-time flight map from OpenSky data"""
        if not flight_data or 'states' not in flight_data:
            print(" No real-time flight data available")
            return None

        # Create world map
        m = folium.Map(location=[20, 0], zoom_start=2)

        flights_added = 0
        for flight in flight_data['states']:
            if flight and len(flight) >= 10:
                callsign = flight[1] or "Unknown"
                latitude = flight[6]
                longitude = flight[5]
                altitude = flight[7] or 0
                velocity = flight[9] or 0

                if latitude and longitude and abs(latitude) <= 90 and abs(longitude) <= 180:
                    # Add plane marker
                    folium.Marker(
                        [latitude, longitude],
                        popup=f"""
                        截 Flight: {callsign}<br>
                         Altitude: {altitude:.0f} m<br>
                         Speed: {velocity:.0f} m/s<br>
                         Position: {latitude:.4f}, {longitude:.4f}
                        """,
                        tooltip=f"Flight: {callsign}",
                        icon=folium.Icon(color='blue', icon='plane', prefix='fa')
                    ).add_to(m)

                    flights_added += 1

                    # Limit for performance
                    if flights_added >= 100:
                        break

        print(f" Added {flights_added} flights to the map")
        return m

def display_map_in_colab(folium_map: folium.Map, width: int = 800, height: int = 600):
    """Display Folium map in Google Colab"""
    if folium_map:
        # Save to HTML and display
        map_html = folium_map._repr_html_()
        display(HTML(f'<div style="width:{width}px; height:{height}px">{map_html}</div>'))
    else:
        print(" No map to display")

def demo_with_sample_data():
    """Demo function that shows sample flight data with map"""
    print("\n" + ""*30)
    print("DEMO MODE - SAMPLE FLIGHT DATA WITH MAP")
    print(""*30)

    sample_flights = [
        {
            'flight_number': 'UA123',
            'airline': 'United Airlines',
            'departure_airport': 'San Francisco International',
            'departure_iata': 'SFO',
            'arrival_airport': 'John F Kennedy International',
            'arrival_iata': 'JFK',
            'status': 'active',
            'scheduled_departure': '2024-01-15T10:00:00+00:00',
            'estimated_departure': '2024-01-15T10:15:00+00:00',
            'scheduled_arrival': '2024-01-15T18:30:00+00:00',
            'estimated_arrival': '2024-01-15T18:45:00+00:00'
        },
        {
            'flight_number': 'AA456',
            'airline': 'American Airlines',
            'departure_airport': 'Los Angeles International',
            'departure_iata': 'LAX',
            'arrival_airport': 'Miami International',
            'arrival_iata': 'MIA',
            'status': 'scheduled',
            'scheduled_departure': '2024-01-15T11:00:00+00:00',
            'estimated_departure': '2024-01-15T11:20:00+00:00',
            'scheduled_arrival': '2024-01-15T19:30:00+00:00',
            'estimated_arrival': '2024-01-15T19:25:00+00:00'
        },
        {
            'flight_number': 'DL789',
            'airline': 'Delta Airlines',
            'departure_airport': 'Atlanta International',
            'departure_iata': 'ATL',
            'arrival_airport': 'London Heathrow',
            'arrival_iata': 'LHR',
            'status': 'active',
            'scheduled_departure': '2024-01-15T12:00:00+00:00',
            'estimated_departure': '2024-01-15T12:10:00+00:00',
            'scheduled_arrival': '2024-01-15T23:30:00+00:00',
            'estimated_arrival': '2024-01-15T23:45:00+00:00'
        }
    ]

    # Display flight info
    for flight in sample_flights:
        print("\n" + "="*60)
        print(f"截  Flight: {flight['flight_number']}")
        print(f" Airline: {flight['airline']}")
        print(f" From: {flight['departure_airport']} ({flight['departure_iata']})")
        print(f" To: {flight['arrival_airport']} ({flight['arrival_iata']})")
        print(f" Status: {flight['status']}")

    # Create and display map
    visualizer = FlightMapVisualizer()
    map_obj = visualizer.create_flight_path_map(sample_flights)

    if map_obj:
        print("\n븝  Displaying flight paths on map...")
        display_map_in_colab(map_obj, width=900, height=600)

    return sample_flights

def get_free_api_guide():
    """Display guide for getting free API keys"""
    print("\n" + ""*40)
    print("HOW TO GET FREE API KEYS:")
    print(""*40)
    print("1.  AviationStack (Recommended):")
    print("    Visit: https://aviationstack.com/")
    print("    Sign up for free account")
    print("    Get 500 requests/month free")
    print("    Perfect for this application")
    print("\n2.  OpenSky Network:")
    print("    No API key needed!")
    print("    Limited but real-time data")
    print("    Great for testing")
    print("\n Tip: Start with OpenSky for testing, then get AviationStack for more features!")

def main_colab():
    """Main function optimized for Google Colab with maps"""

    print(""*50)
    print("           GOOGLE COLAB FLIGHT TRACKER WITH MAPS")
    print(""*50)

    # Show API guide first
    get_free_api_guide()

    # Initialize trackers and visualizer
    opensky_tracker = OpenSkyTracker()
    map_visualizer = FlightMapVisualizer()

    # Ask if user wants to use AviationStack
    use_aviationstack = input("\nDo you have an AviationStack API key? (yes/no): ").strip().lower()

    tracker = None
    if use_aviationstack in ['yes', 'y']:
        api_key = getpass.getpass("Enter your AviationStack API key (hidden): ")
        if api_key:
            tracker = FlightTracker(api_key)
            print(" AviationStack tracker initialized!")
        else:
            print(" No API key provided. Using OpenSky Network only.")
    else:
        print(" Using OpenSky Network (no API key required)")

    while True:
        print("\n" + "="*60)
        print(" LIVE FLIGHT TRACKER WITH MAPS")
        print("="*60)

        if tracker:
            print("1.  Search by Flight Number + Show Map")
            print("2.  Search by Airline + Show Map")
            print("3.  Departures from Airport + Show Map")
            print("4.  Arrivals at Airport + Show Map")
            print("5.  OpenSky - Real-time Flight Map")
            print("6.  View Sample Data with Map")
            print("7.  Exit")
        else:
            print("1.  OpenSky - Real-time Flight Map")
            print("2.  View Sample Data with Map")
            print("3.  Exit")

        print("="*60)

        if tracker:
            choice = input("Enter your choice (1-7): ").strip()
        else:
            choice = input("Enter your choice (1-3): ").strip()
            # Map choices for OpenSky-only mode
            choice_map = {'1': '5', '2': '6', '3': '7'}
            choice = choice_map.get(choice, choice)

        if choice == '1' and tracker:
            flight_number = input("Enter flight number (e.g., UA123): ").strip().upper()
            print(f" Searching for flight {flight_number}...")
            flight_data = tracker.get_flights_by_flight_number(flight_number)
            flights_info = tracker.display_flight_info(flight_data)

            if flights_info:
                print("\n븝  Generating flight path map...")
                map_obj = map_visualizer.create_flight_path_map(flights_info)
                display_map_in_colab(map_obj, width=900, height=600)

        elif choice == '2' and tracker:
            airline = input("Enter airline IATA code (e.g., UA): ").strip().upper()
            print(f" Searching for {airline} flights...")
            flight_data = tracker.get_flights_by_airline(airline)
            flights_info = tracker.display_flight_info(flight_data)

            if flights_info:
                print("\n븝  Generating flight paths map...")
                map_obj = map_visualizer.create_flight_path_map(flights_info)
                display_map_in_colab(map_obj, width=900, height=600)

        elif choice == '3' and tracker:
            airport = input("Enter departure airport IATA code (e.g., JFK): ").strip().upper()
            print(f" Searching departures from {airport}...")
            flight_data = tracker.get_flights_by_airport(airport, "departure")
            flights_info = tracker.display_flight_info(flight_data)

            if flights_info:
                print("\n븝  Generating departure routes map...")
                map_obj = map_visualizer.create_flight_path_map(flights_info)
                display_map_in_colab(map_obj, width=900, height=600)

        elif choice == '4' and tracker:
            airport = input("Enter arrival airport IATA code (e.g., LAX): ").strip().upper()
            print(f" Searching arrivals at {airport}...")
            flight_data = tracker.get_flights_by_airport(airport, "arrival")
            flights_info = tracker.display_flight_info(flight_data)

            if flights_info:
                print("\n븝  Generating arrival routes map...")
                map_obj = map_visualizer.create_flight_path_map(flights_info)
                display_map_in_colab(map_obj, width=900, height=600)

        elif choice == '5':
            print(" Fetching real-time flights from OpenSky Network...")
            flight_data = opensky_tracker.get_all_flights()

            if flight_data and 'states' in flight_data:
                print(f" Found {len(flight_data['states'])} current flights")
                print("븝  Generating real-time flight map...")
                map_obj = map_visualizer.create_realtime_flight_map(flight_data)
                display_map_in_colab(map_obj, width=900, height=600)
            else:
                print(" No real-time flight data available")

        elif choice == '6':
            demo_with_sample_data()

        elif choice == '7':
            print(" Thank you for using Google Colab Flight Tracker with Maps!")
            break

        else:
            print(" Invalid choice. Please try again.")

        # Add a small delay to avoid API rate limits
        time.sleep(2)

# Run the application
if __name__ == "__main__":
    print("\n Choose an option:")
    print("1.  Run Full Flight Tracker with Maps")
    print("2. ㎦ Quick Test (Sample Data with Map)")

    option = input("Enter choice (1 or 2): ").strip()

    if option == '1':
        main_colab()
    elif option == '2':
        demo_with_sample_data()
    else:
        print(" Invalid choice. Running full tracker...")
        main_colab()
```
